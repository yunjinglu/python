```python
import numpy as np
import pandas as pd

frame = pd.read_excel('學生成績單.xlsx')
frame
```

```python
students_frame = frame[["國文","英文","數學","地理","歷史"]]
students_frame
```

```python
#mean(),std(),max(),min(),median()中位數,corr()相關係數
students_frame['國文'].mean()
```

```python
students_frame.mean()
```

```python
students_frame.mean(axis=1) #axis=1算一列的平均 預設axis=0是算一行的平均
```

```python
display(students_frame['國文'])
for item in students_frame['國文']:
    if item >= 80:
        print('優')
    elif item >= 70:
        print('甲')
    elif item >= 60:
        print('乙')
    else:
        print('丙')
```

```python
def fun1(item):
    if item >= 80:
        return '優'
    elif item >= 70:
        return '甲'
    elif item >= 60:
        return '乙'
    else:
        return '丙'
    
students_frame['國文'].map(fun1)
```

```python
students_frame['國文等級'] = students_frame['國文'].map(fun1)
students_frame
```

```python
students_frame = frame[["國文","英文","數學","地理","歷史"]]
students_frame
```

```python
def fun1(item):
    if item >= 80:
        return '優'
    elif item >= 70:
        return '甲'
    elif item >= 60:
        return '乙'
    else:
        return '丙'
students_frame.map(fun1)
```

```python
def fun2(s):
    return s.min()

students_frame.apply(fun2)
```

```python
students_frame.apply(fun2,axis=1)
```

```python
def fun3(s):
    return s[s < 60].count() #count()計算個數
students_frame.apply(fun3,axis=1)
```

```python
students_frame['不及格科目數量'] = students_frame.apply(fun3,axis=1)
students_frame
```

```python
def fun5(s):
    return pd.Series([s.max(),s.min(),s.median(),s.mean(),s.std()],
                     index=["最高分","最低分","中間值","平均","標準差"])
students_frame.apply(fun5)
```
