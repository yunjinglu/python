```python
import csv
with open("個股日成交資訊.csv",encoding='utf-8',newline='') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)
    all_list = []
    for item in csv_reader:
        all_list.append([item[0], item[1], item[2], item[3], item[7], item[8], item[9]])

all_list
```
