```python
import pandas as pd

frame = pd.read_csv('world.csv')
frame.info() #檢查資料
```

```python
world_frame = frame[['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數']]
world_frame
```

```python
#subscript的搜尋
asia_frame = world_frame[world_frame["洲名"] == "亞洲"]
asia_frame
```

```python
#subscript的搜尋
condition1 = (world_frame["國家"] == "台灣") & (world_frame["新增確診數"] == 0) #要加() &代表and |代表or
condition1
```

```python
world_frame[condition1]
```

```python
#使用query('')的搜尋
world_frame.query('國家=="台灣"')
```

```python
world_frame.query('國家=="台灣" and 新增確診數==0')
```

```python
world_frame.query('日期=="2022-05-13"')
```

```python
world_frame.query('日期=="2022-05-13" and 新增確診數>50')
```

```python
world_frame.query('國家=="台灣" and 日期>="2021-01-01" and 日期<="2021-01-31"')
```

```python
taiwan = world_frame.query('國家=="台灣"')
taiwan
```

```python
taiwan_max = taiwan['新增確診數'].max()
taiwan_max
```

```python
taiwan.query('新增確診數==@taiwan_max') #變數前要加@
```

```python
japan = world_frame.query('國家=="日本"')
japan_max = japan['新增確診數'].max()
japan.query('新增確診數==@japan_max')
```
