```python
import pandas as pd

df = pd.read_csv("world.csv")
allDataFrame = df[['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數']]
taiwan = allDataFrame[allDataFrame["國家"] == "台灣"]
taiwan
```

```python
#新增確診數>5萬人
taiwan[taiwan["新增確診數"] >= 50000]
```

```python
#取出最多新增確診數
max = taiwan["新增確診數"].max()
taiwan[taiwan['新增確診數'] == max]
```
